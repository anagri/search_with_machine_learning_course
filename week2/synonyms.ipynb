```python
import fasttext
```

```python
# Train model
model = fasttext.load_model("/workspace/search_with_machine_learning_course/title_model.bin")
```

```python
model.get_nearest_neighbors('iphone')
```

```python
synonyms = []
```

```python
min_threshold = 0.75
```

```python
with open('/workspace/datasets/fasttext/top_words.txt') as f:
    for line in f:
        nns = model.get_nearest_neighbors(line)
        syns = [nn for (threshold, nn) in nns if float(threshold) >= min_threshold]
        if len(syns) > 1:
            synonyms.append(syns)
```

```python
synonyms
```

```python
outfile = '/workspace/datasets/fasttext/synonyms.csv'
```

```python
with open(outfile, 'w') as f:
    for synonym in synonyms:
        f.write("{}\n".format(",".join(synonym)))
```
